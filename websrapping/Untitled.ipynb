```python
import pandas as pd
```

```python
import requests
import bs4
```

```python
url="https://www.vibrantacademy.com/Vibrant/contactus"
```

```python
resp=requests.get(url)
```

```python
resp.status_code
```

```python
resp.reason
```

```python
soup=bs4.BeautifulSoup(resp.content,'html5lib')
```

```python
soup
```

```python
divs=soup.find_all('div',attrs={'class':'text pt-1'})
```

```python
divs[0]
```

```python
divs[0].find('h4').text
```

```python
center=[]
office=[]
```

```python
for div in divs:
    center.append(div.find('h4').text)
    of=div.find('span').text
    office.append(of[:-1])
```

```python
center
```

```python
office
```

```python
loca=[]
for i in divs:
    loc=i.div.div.p.text
    loca.append('\n')
    loca.append(loc)
    print(loc)
```

```python
contact=[]
for i in divs:
    loc=i.div.div.a.text
    contact.append(loc)
```
