```python
import tqdm
```

```python
def get_content(START, END):
    e_page = []
    page = 0
    url = "https://www.flipkart.com/search?q=laptops&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_4_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_4_0_na_na_na&as-pos=4&as-type=TRENDING&suggestionId=laptops&requestId=58d152c1-2bb3-4681-a0a5-ddfdf904c824&page={}"
    data = {
            'Product Name': [],
            'Price': [],
            'Customer Name': [],
            'Rating': [],
            'Reviews': []
            }
    for c in tqdm.tqdm(range(START, END+1)):
        resp = requests.get(url.format(c, c))
        if resp.status_code == 200:
            context = resp.text
            soup = BeautifulSoup(context, 'html.parser')
            vars1 = []    #Names
            names = []
            names = [ tags.div.text.split('.')[1] for tags in soup.find_all('div', attrs={'class': '_1W9f5C'})]
            for i in names:
                var = i.split('-')[0]
                vars1.append(var)
            data['Product Name'].extend(vars1)
            prices = [ tags.text for tags in soup.find_all('div', attrs={'class': '_30jeq3 UMT9wN'})]  #Prices
            data['Price'].extend(prices)
            custom_name = [ tags.text for tags in soup.find_all('p', attrs={'class': '_2sc7ZR _2V5EHH'})]  #Cus_Name
            data['Customer Name'].extend(custom_name)
            ratings = [ tags.text for tags in soup.find_all('div', attrs={'class': '_3LWZlK _1BLPMq'})]   #Ratings
            data['Rating'].extend(ratings)
            reviews = [ tags.div.text for tags in soup.find_all('div', attrs={'class': 't-ZTKy'})]   #Reviews
            data['Reviews'].extend(reviews)
        elif resp.status_code == 503:
            e_page.append(c)
        else:
            print(f"!Error !! Status Code {resp.status_codetatus_code} and reason {resp.reason}.")
            break
    page += 1    
    return data
```
